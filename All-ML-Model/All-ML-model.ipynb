```python

from google.colab import drive
drive.mount('/content/drive')


import os, warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    RocCurveDisplay,
    PrecisionRecallDisplay,
    roc_auc_score
)



CSV_PATH = "/content/drive/MyDrive/ML/Breast_Cancer_Diagnostic.csv"
df = pd.read_csv(CSV_PATH)


display(df.head())
display(df.info())
display(df.describe())
display(df.isnull().sum())


x = df.drop("diagnosis", axis=1)
y = df["diagnosis"]

 from sklearn.model_selection import train_test_split
 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.30, random_state=40)
 
 from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(x_train, y_train)
model.score(x_test,y_test)

from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
y_train_Encode= le.fit_transform(y_train)
y_test_Encode= le.fit_transform(y_test)


from sklearn.linear_model import LinearRegression
model1= LinearRegression()
model1.fit(x_train, y_train_Encode)
model1.score(x_test, y_test_Encode)

from sklearn.tree import DecisionTreeClassifier
model2= DecisionTreeClassifier()
model2.fit(x_train, y_train)
model2.score(x_test,y_test)


from sklearn.svm import SVC
model3= SVC(kernel='linear',C=30,gamma='auto')
model3.fit(x_train, y_train)
model3.score(x_test,y_test)

from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=100, random_state=40)
model_rf.fit(x_train, y_train)
print("Random Forest Score:", model_rf.score(x_test, y_test))


from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(x_train, y_train)
print("KNN Score:", model_knn.score(x_test, y_test))

from sklearn.naive_bayes import GaussianNB
model_nb = GaussianNB()
model_nb.fit(x_train, y_train)
print("Naive Bayes Score:", model_nb.score(x_test, y_test))

from sklearn.ensemble import GradientBoostingClassifier
model_gbm = GradientBoostingClassifier(n_estimators=100, random_state=40)
model_gbm.fit(x_train, y_train)
print("Gradient Boosting Score:", model_gbm.score(x_test, y_test))

from sklearn.ensemble import AdaBoostClassifier
model_adaboost = AdaBoostClassifier(n_estimators=100, random_state=40)
model_adaboost.fit(x_train, y_train)
print("AdaBoost Score:", model_adaboost.score(x_test, y_test))

import xgboost as xgb
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)

model_xgb = xgb.XGBClassifier(objective='binary:logistic', n_estimators=100, random_state=40)
model_xgb.fit(x_train, y_train_encoded)
print("XGBoost Score:", model_xgb.score(x_test, y_test_encoded))

import lightgbm as lgb
from sklearn.preprocessing import LabelEncoder


if y_train.dtype == 'object':
    le = LabelEncoder()
    y_train_encoded = le.fit_transform(y_train)
    y_test_encoded = le.transform(y_test)
else:
    y_train_encoded = y_train
    y_test_encoded = y_test


model_lgbm = lgb.LGBMClassifier(n_estimators=100, random_state=40)
model_lgbm.fit(x_train, y_train_encoded)
print("LightGBM Score:", model_lgbm.score(x_test, y_test_encoded))
!pip install catboost
from catboost import CatBoostClassifier
from sklearn.preprocessing import LabelEncoder


if y_train.dtype == 'object':
    le = LabelEncoder()
    y_train_encoded = le.fit_transform(y_train)
    y_test_encoded = le.transform(y_test)
else:
    y_train_encoded = y_train
    y_test_encoded = y_test

model_catboost = CatBoostClassifier(iterations=100, random_state=40, verbose=0)
model_catboost.fit(x_train, y_train_encoded)
print("CatBoost Score:", model_catboost.score(x_test, y_test_encoded))

model_scores = {
    "Logistic Regression": model.score(x_test, y_test),
    "Decision Tree": model2.score(x_test, y_test),
    "SVC (Linear)": model3.score(x_test, y_test),
    "Random Forest": model_rf.score(x_test, y_test),
    "KNN": model_knn.score(x_test, y_test),
    "Naive Bayes": model_nb.score(x_test, y_test),
    "Gradient Boosting": model_gbm.score(x_test, y_test),
    "AdaBoost": model_adaboost.score(x_test, y_test),
    "XGBoost": model_xgb.score(x_test, y_test_encoded), 
    "LightGBM": model_lgbm.score(x_test, y_test_encoded), 
    "CatBoost": model_catboost.score(x_test, y_test_encoded) 
}

models = list(model_scores.keys())
scores = list(model_scores.values())

plt.figure(figsize=(12, 6))
sns.barplot(x=models, y=scores)
plt.xticks(rotation=45, ha='right')
plt.ylabel("Accuracy Score")
plt.title("Comparison of Model Accuracies")
plt.tight_layout()
plt.show()

```
